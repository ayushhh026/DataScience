```python
class Person:
    def __init__(self,name,age):
        self.__name=name
        self.__age=age
    

    #getter method
    def get_name(self):
        return self.__name
    
    #setter method to change private variable
    def set_name(self,name):
        self.__name=name
```

```python
person =Person("Spongebob",30)
```

```python
print(person.get_name())
```

```python
print(person.set_name("patrick"))
```
