```python
import matplotlib.pyplot as plt
```

```python
x=[1,2,3,4,5]
y=[1,4,9,16,25]
```

```python
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.title("Basic line plot")
plt.show()

plt.plot(x,y,color="red",linestyle='-',marker='o')
```

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[1,2,3,4,5]

plt.figure(figsize=(9,5))

plt.subplot(1,2,1)

plt.plot(x,y,color='green')
```

```python
###BAR PLOT
categories=['A','B','C','D','E']
values=[5,7,3,8,6]

plt.bar(categories,values,color='red')
```

```python
##HISTOGRAM
data=[1,1,2,3,2,2,1,3,4,5,6,6,4,5,5,6,5,4,4,4,5,5,5,5,3,2,2,2,3,1,2,1,2]
plt.hist(data,bins=6,color='Orange',edgecolor='black')
```

```python
###SCATTER PLOT 
x=[1,2,3,4,5]
y=[2,3,4,5,6]
plt.scatter(x,y,color='red',marker='x')
```

```python
### pie chart
labels=['A','B','C','D']
size=[30,10,20,40]
colors=['gold','orange','green','red']
explode=(0.2,0,0,0)

plt.pie(size,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
```

```python
import pandas as pd
sales_data=pd.read_csv('pokemon.csv')
```

```python
sales_data
```

```python
## plot total sales by product 
total=sales_data.groupby('Type1')['Weight'].sum()
```

```python
print(total)
```

```python
total.plot(kind='pie')
```

```python
#height to weight ratio 
hw=sales_data.groupby('Height')['Weight'].sum().reset_index()
print(hw)
```

```python
hw.plot(kind='bar')
```
