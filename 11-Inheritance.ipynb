```python
class Car:
    def __init__(self,windows,doors,engine_type):
        self.windows=windows
        self.doors=doors
        self.engine_type=engine_type
    def drive(self):
        print(f"The person will drive {self.engine_type} car")
c1=Car(3,4,'Petrol')        
```

```python
c1.drive()
```

```python
c1.drive
```

```python
class Tesla(Car):
    def __init__(self, windows, doors, engine_type,is_selfdrive):
        super().__init__(windows, doors, engine_type)
        self.is_selfdrive=is_selfdrive
    def self_drive(self):
        print(f"Tesla is a {self.is_selfdrive} {self.engine_type} car")
    
```

```python
Tesla1=Tesla(3,4,'Electric',True)
```

```python
Tesla1.self_drive()
```

```python
Tesla1.drive()
```
