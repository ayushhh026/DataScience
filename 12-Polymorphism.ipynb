```python
#USING METHOD OVERRIDING 
class Shape:
    def area(self):
        return "The area of figure"

class Rectangle(Shape):
    def __init__(self,width,height):
        self.width=width
        self.height=height
    def area(self):
        return self.width*self.height
    
class Circle(Shape):
    def __init__(self,radius):
        self.radius=radius
    def area(self):
        return 3.142 * self.radius**2

def print_area(shape):
    print(f"The area is {shape.area()}")
```

```python
rect=Rectangle(10,4)
cir=Circle(2)
```

```python
print_area(rect)
print_area(cir)
```

```python
sha=Shape()
```

```python
print_area(sha)
```
