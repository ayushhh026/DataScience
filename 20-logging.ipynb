```python
#MULTIPLE LOGGERSS

## CREATE A LOGGER FOR ONLY MODULE 1

logger1=logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

## CREATE A LOGGER FOR ONLY MODULE 2

logger2=logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

logging.basicConfig(filename='app.log',filemode='w', level=logging.DEBUG,format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
                    datefmt='%Y-%m-%d %H:%M:S')
```

```python
### LOG message with different loggers
logger1.debug("This is a debug message for module 1")
logger1.warning("This is a debug message for module 2")
```

```python
import logging
logging.basicConfig(level=logging.DEBUG,format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
                    datefmt='%Y-%m-%d %H:%M:S',handlers=[
                        logging.FileHandler("app1.log"),
                        logging.StreamHandler()
                    ])
```

```python
logger=logging.getLogger("ArithmethicApp")

def add(a,b):
    result=a+b 
    logger.debug(f"Adding {a} and {b} result ={result} ")
    return result
def mul(a,b):
    result=a*b 
    logger.warning(f"multiply {a} and {b} result ={result} ")
    return result
```

```python
add(10,15)
mul(20,10)
```
