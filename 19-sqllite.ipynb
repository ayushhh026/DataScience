```python
import sqlite3
```

```python
## Connect and create a db 
connection=sqlite3.connect('example.db')
```

```python
cursor=connection.cursor()
```

```python
## Create a table 
cursor.execute('''
CREATE TABLE IF NOT EXISTS employee(
    id INT primary key,
    name text NOT NULL,
    age INT,
    department text
    )
''')

##Commit the changs 
connection.commit()
```

```python
cursor.execute('''
SELECT * FROM employee
''')
```

```python
## INSERT DATA IN SQL LITE 3 DB
cursor.execute(
    '''
    INSERT INTO employee(name,age,department)
            values('Spongebob',30,'Data Scientist')
'''
) 
cursor.execute(
    '''
    INSERT INTO employee(name,age,department)
            values('patrick',30,'Data Engineer')
'''
) 
cursor.execute(
    '''
    INSERT INTO employee(name,age,department)
            values('Eugen',30,'Cashier')
'''
) 

connection.commit()
```

```python
### QUERRY DATA FROM SQL LITE

cursor.execute ('Select * FROM employee')
rows=cursor.fetchall()

for row in rows:
    print(row)
```

```python
### Update data in the tabnle 

cursor.execute(
    '''
    Update employee
                SET age=34 WHERE name='Spongebob'
'''
)
connection.commit()
```

```python
## DELETE DATA FROM TABLE 

cursor.execute(''' 
DELETE FROM employee
WHERE name='Spongebob'              
               ''')
connection.commit()
```
