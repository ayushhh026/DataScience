```python
class BankAccount:
    def __init__(self,name,balance):
        self.name=name 
        self.balance=balance
    def deposit(self,amount):
        self.balance+=amount
        print(f"You have successfully deposited {amount} and your current balance is {self.balance}")
    
    def withdraw(self,amount):
        if(amount>self.balance):
            print("Insufficient funds")
        else:
            self.balance-=amount
            print(f"you have sucessfully withdrawn {amount} and your current balance is {self.balance}")
    
    def balance(self):
        return self.balance

account=BankAccount("Spongebob",5000)
```

```python
print(account.balance)
```

```python
print(account.deposit(12345))
```

```python
print(account.withdraw(12344))
```

```python
print(account.balance)
```
