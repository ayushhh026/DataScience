```python
## BASIC PLOTTING
tips=sns.load_dataset('tips')
tips
```

```python
## Scatter plot 
import matplotlib.pyplot as plt
```

```python
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title(("Scatter plot"))
plt.show()
```

```python
#Line plot 
sns.lineplot(x='size',y='total_bill',data=tips)
```

```python
##categorial plots 
#bar plot
sns.barplot(x='day',y='total_bill',data=tips)
```

```python
#Box plot 
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
##violent plot 
sns.violinplot(x='day',y='total_bill',data=tips)
```

```python
### histograms 
sns.histplot(tips['total_bill'],bins=5,kde=True) #kerneldensity indicator which shows the shape
```

```python
### KDE plots 
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
# Pair plot 
sns.pairplot(tips) # gives all the relation
```

```python
## Heat map 
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr)
```

```python
sns.heatmap(corr,annot=True,cmap='coolwarm')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

x = [1,2,3,4]
y = [10,20,25,30]

plt.plot(x, y)
plt.title("Line Plot")
plt.show()
names = ["A","B","C"]
values = [10,20,15]

plt.bar(names, values)
plt.title("Bar Chart")
plt.show()
data = [1,2,2,3,3,3,4,4,4,4]

plt.hist(data)
plt.title("Histogram")
plt.show()

sizes = [30,40,30]
labels = ["A","B","C"]

plt.pie(sizes, labels=labels)
plt.title("Pie Chart")
plt.show()

x = [1,2,3,4]
y = [10,15,13,17]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.show()

data = [10,20,30,40,50]

plt.boxplot(data)
plt.title("Box Plot")
plt.show()


data = np.random.rand(5,5)

sns.heatmap(data)
plt.title("Heatmap")
plt.show()

x = [1,2,3,4]
y = [10,20,15,25]

plt.fill_between(x, y)
plt.title("Area Plot")
plt.show()

data = [10, 20, 30, 40, 50, 60, 70]

sns.violinplot(data=data)
plt.title("Violin Plot")
plt.show()

import seaborn as sns
import matplotlib.pyplot as plt

data = [10,20,20,30,30,30,40,50]

sns.kdeplot(data)
plt.title("Density Plot")
plt.show()
```
